```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
sigma = 10
rho = 28
beta = 8/3

# Lorenz system
def lorenz(x, y, z, s=sigma, r=rho, b=beta):
    dx = s * (y - x)
    dy = x * (r - z) - y
    dz = x * y - b * z
    return dx, dy, dz

# Time parameters
dt = 0.01
num_steps = 10000

# Initial values
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)
xs[0], ys[0], zs[0] = (0., 1., 1.05)

# Integrate the Lorenz equations
for i in range(num_steps):
    dx, dy, dz = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + dx * dt
    ys[i + 1] = ys[i] + dy * dt
    zs[i + 1] = zs[i] + dz * dt

# Plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, ys, zs, lw=0.5, color='k')
ax.set_title(f'Lorenz: $\\sigma = {sigma}, \\beta = {beta:.3f}, \\rho = {rho}$')
plt.savefig('lorenz.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
a = 0.1
b = 0.1
c = 18

cut_tail = 5000

# Roessler system
def roessler(x, y, z, a=a, b=b, c=c):
    dx = -y - z
    dy = x + a * y
    dz = b + z * (x - c)
    return dx, dy, dz

# Time parameters
dt = 0.01
num_steps = 20000

# Initial values
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)
xs[0], ys[0], zs[0] = (1, 0, 0)

# Integrate the Roessler equations
for i in range(num_steps):
    dx, dy, dz = roessler(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + dx * dt
    ys[i + 1] = ys[i] + dy * dt
    zs[i + 1] = zs[i] + dz * dt

# Plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs[cut_tail:], ys[cut_tail:], zs[cut_tail:], lw=0.5)
ax.set_title(f'Roessler: a={a}, b={b}, c={c}')
plt.savefig('roessler.png')
```
